```python
from pyspark.sql.window import Window
from pyspark.sql.types import StructType,StructField, StringType, IntegerType
import pyspark.sql.functions as F
```

```python
def spark_fe(df,fe_str,spec_udf):

    # spec_udf = mean_udf
    # fe_str = "mean"
    # df.withColumn(fe_str+"_s1",spec_udf("s1").over(windowSpec)) => memory error
    
    windowSpec = Window().partitionBy(['id']).orderBy('cycle').rowsBetween(-6,0)
    
    df_Roll = df.select(df.columns+
        [spec_udf("s1").over(windowSpec).alias(fe_str+"_s1"),
         spec_udf("s2").over(windowSpec).alias(fe_str+"_s2"),
         spec_udf("s3").over(windowSpec).alias(fe_str+"_s3"),
         spec_udf("s4").over(windowSpec).alias(fe_str+"_s4"),
         spec_udf("s5").over(windowSpec).alias(fe_str+"_s5"),
         spec_udf("s6").over(windowSpec).alias(fe_str+"_s6"),
         spec_udf("s7").over(windowSpec).alias(fe_str+"_s7"),
         spec_udf("s8").over(windowSpec).alias(fe_str+"_s8"),
         spec_udf("s9").over(windowSpec).alias(fe_str+"_s9"),
         spec_udf("s10").over(windowSpec).alias(fe_str+"_s10"),
         spec_udf("s11").over(windowSpec).alias(fe_str+"_s11"),
         spec_udf("s12").over(windowSpec).alias(fe_str+"_s12"),
         spec_udf("s13").over(windowSpec).alias(fe_str+"_s13"),
         spec_udf("s14").over(windowSpec).alias(fe_str+"_s14"),
         spec_udf("s15").over(windowSpec).alias(fe_str+"_s15"),
         spec_udf("s16").over(windowSpec).alias(fe_str+"_s16"),
         spec_udf("s17").over(windowSpec).alias(fe_str+"_s17"),
         spec_udf("s18").over(windowSpec).alias(fe_str+"_s18"),
         spec_udf("s19").over(windowSpec).alias(fe_str+"_s19"),
         spec_udf("s20").over(windowSpec).alias(fe_str+"_s20"),
         spec_udf("s21").over(windowSpec).alias(fe_str+"_s21")
        ])
    
    return df_Roll
```

```python
@F.pandas_udf("double")
def av_udf(v: pd.Series) -> float:
    return v.mean()

@F.pandas_udf("double")
def sd_udf(v: pd.Series) -> float:
    return v.std()

@F.pandas_udf("double")
def pk_udf(v: pd.Series) -> float:
    return v.max()

@F.pandas_udf("double")
def rms_udf(v: pd.Series) -> float:
    return np.sqrt((v**2).sum()/len(v))

@F.pandas_udf("double")
def skew_udf(v: pd.Series) -> float:
    return v.skew()

@F.pandas_udf("double")
def kurt_udf(v: pd.Series) -> float:
    return v.kurt()

@F.pandas_udf("double")
def crest_udf(v: pd.Series) -> float:
    rms_sep = np.sqrt((v**2).sum()/len(v))
    return np.abs(v.max()) / rms_sep

@F.pandas_udf("double")
def clear_udf(v: pd.Series) -> float:
    res_sep = np.square ((v**0.5).sum()/len(v))
    return np.abs(v.max()) / res_sep

@F.pandas_udf("double")
def shape_udf(v: pd.Series) -> float:
    rms_sep = np.sqrt((v**2).sum()/len(v))
    return rms_sep / v.mean()

@F.pandas_udf("double")
def imp_udf(v: pd.Series) -> float:
    return v.max() / v.mean()
```

```python
fe_str_list = ["av","sd","pk","rms","skew","kurt","crest","clear","shape","imp"]
fe_udf_list = [av_udf,sd_udf,pk_udf,rms_udf,skew_udf,kurt_udf,crest_udf,clear_udf,shape_udf,imp_udf]
```

```python
df_fe = df.alias("df_fe")

for i in tqdm(range(len(fe_str_list))):
    
    df_fe = spark_fe(df_fe, fe_str_list[i], fe_udf_list[i])
```
